```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
from langchain_openai import OpenAIEmbeddings
from langchain.vectorstores import chroma
```

```python
text = "This is a practice session for LanchChain and ChromaDB."
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
query_results=embeddings.embed_query(text)
query_results
```

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader("speech.txt")
docs = loader.load()
docs

from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
texts = text_splitter.split_documents(docs)
texts
```

```python
# Vector Embedding and Vector Store DB
from langchain.vectorstores import Chroma

db= Chroma.from_documents(texts, emb_1024, )
db
```

```python
## Retrieving the results from ChromaDB
query_text = "without countenance except from a lawless and malignant"

retrieved_results = db.similarity_search(query_text)
print(retrieved_results)
```
