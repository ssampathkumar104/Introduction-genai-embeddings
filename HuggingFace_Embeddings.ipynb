#### Sentence Transformers

```python
os.environ["HF_TOKEN"] = os.getenv("HF_TOKEN")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text = "This is a test sentence for embeddings."
hf_query_result = embeddings.embed_query(text)
print(hf_query_result)
```

```python
len(hf_query_result)  # Should print the length of the embedding vector
```

```python
hf_doc_result = embeddings.embed_documents([text,"This is not a test document."])
hf_doc_result[0]
```
