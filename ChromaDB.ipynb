```python
from langchain_community.embeddings import OllamaEmbeddings
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader = TextLoader("speech.txt")
docs = loader.load()
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=20)
texts = splitter.split_documents(docs)
texts
```

```python
embeddings = OllamaEmbeddings(model="llama2")
db = Chroma.from_documents(texts, embeddings)
db
```

```python
query="How many times did the speaker mention the word 'freedom'?"
docs = db.similarity_search(query)
docs[0].page_content
```

```python
#Saving

vectordb = Chroma.from_documents(documents=texts,embedding=embeddings,persist_directory="/.chroma_db")

```

```python
db2 = Chroma(persist_directory="/.chrome_db",embedding_function=embeddings)
docs2 = db2.similarity_search(query)
docs2
```
